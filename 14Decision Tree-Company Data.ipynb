```python
df['Sales'] = pd.cut(df.Sales,bins=[0,6,12,17], labels=['Low','Medium', 'High'], right = False)
```

```python
#converting categorical data into numeric 
from sklearn.preprocessing import LabelEncoder
Encoder=LabelEncoder()
```

```python
df['Sales']=Encoder.fit_transform(df["Sales"])
df['ShelveLoc']=Encoder.fit_transform(df["ShelveLoc"])
df['Urban']=Encoder.fit_transform(df["Urban"])
df['US']=Encoder.fit_transform(df["US"])
```

```python
df
```

```python
df["Sales"].unique()
```

```python
#dividing data into x & y variable 
x=df.iloc[:,1:11]

y=df['Sales']
```

```python
#spliting data into training and testing
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=23)
```

```python
#Building Decision Tree Classifier using Entropy Criteria

#Type Markdown and LaTeX: ùõº2
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier(criterion="entropy",max_depth=3)
model.fit(x_train,y_train)
```

```python
#plot the decision tree
from sklearn import tree
import  matplotlib.pyplot as  plt
```

```python
#PLot the decision tree
plt.figure(figsize=(18,10))

tree.plot_tree(model);
```

```python
df.columns
```

```python
fig, axes = plt.subplots(figsize = (18,18))
fn=['CompPrice', 'Income', 'Advertising', 'Population', 'Price',
       'ShelveLoc', 'Age', 'Education', 'Urban', 'US']
cn=['Low','Medium', 'High']

tree.plot_tree(model,
               feature_names = fn,
               class_names=cn,
               filled = True,
               fontsize=10);
```

```python
model.feature_importances_
```

```python
# predicting on test data set 
preds = model.predict(x_test) 

# getting the count of each category 
pd.Series(preds).value_counts()
```

```python
pd.crosstab(y_test,preds)
```

```python
# Accuracy 
np.mean(preds==y_test)*100
```

```python
#Building Decision Tree Classifier (CART)using Gini Criteria
```

```python
model_gini=DecisionTreeClassifier(criterion="gini",max_depth=3)
```

```python
model_gini.fit(x_train,y_train)
```

```python
#PLot the decision tree
plt.figure(figsize=(18,10))

tree.plot_tree(model_gini);
```

```python
fig, axes = plt.subplots(figsize = (18,18))
fn=['CompPrice', 'Income', 'Advertising', 'Population', 'Price',
       'ShelveLoc', 'Age', 'Education', 'Urban', 'US']
cn=['Low','Medium', 'High']

tree.plot_tree(model_gini,
               feature_names = fn,
               class_names=cn,
               filled = True,
               fontsize=10);
```
